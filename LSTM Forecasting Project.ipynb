## Baseline Moving Average Model
A moving average model predicts what the next value will be using the current moving average of the past 30 days. 
When predicting on the test set, I used actual test set values to calculate the moving average at any given point after a prediction was made for the day before. 
In a production scenario where forecasting for multiple days was being done and future
values were unknown, you would use the moving average prediction recursively to make predictions one at a time.

```python
# Create series out of train and test dfs to calculate predictions
ma_train = train_df.Close
ma_test = test_df.Close
```

```python
# Make predictions for the test set by  
# making 1 prediction at a time and recalculating MA
# The last value of moving_avg will be used as the forecast for the first value in the test set
# Then, recompute the moving average (including the first test value), and so on
ma_forecasts = []
for t in range(len(ma_test)):
    # Initial forecast = the last MA of train set
    if t == 0:
        forecast = ma_train.iloc[-1] 
    else:
        # Recalculate moving average including known test values
        updated_train = pd.concat([ma_train, ma_test[:t]])
        forecast = updated_train.rolling(window=30).mean().iloc[-1]
    ma_forecasts.append(forecast)
```

```python
test_predictions = pd.DataFrame(ma_forecasts, index=test_df.index, columns=['Prediction'])
```

```python
rmse = np.sqrt(mean_squared_error(test_df, test_predictions))
print(f'{STOCK} RMSE: {rmse:.2f}')
```

## Prophet Time Series Model

Interestingly, Prophet performed much worse than the moving average baseline model. With hyperparameter tuning, Prophet would likely improve. But for this dataset, Prophet did not pick up the patterns very well.
Prophet is generally good at picking up time series that have more consistent seasonality and/or trends. It appears that for this test set, Prophet thought there was going to be a downward trend, but the prices actually went back up after a couple dips. ## LSTM Model

LSTMs are well known for doing well with time series. For the architecture, I started with 2 LSTM layers, each with 50 units, each followed by a 20% dropout rate layer. This model seemed to pick up the time series pattern pretty well and did much better than both MA and Prophet models. I tested out adding an additional LSTM layer, but this made the performance worse.

```python
def lstm_model(X_train,dense=1):
    # Build the LSTM model
    model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(dense)
])
    return model
```

```python
def create_sequences(series,window_size):
    # Initialize lists for X and y data
    X,y = [],[]
    # Create sequences from scaled data
    # Go from 0 to length of series/data - window size to
    # avoid index out of bounds error
    for i in range(len(series) - window_size):
        row = [[x] for x in series[i:i+window_size]]
        X.append(row)
        # Target = label/outcome/y (wherever we are at in index i, go window_size steps ahead
        # to get y / true value)
        target = series[i+window_size]
        y.append(target)

    X=np.array(X)
    y=np.array(y)
    return X,y
```

## Tune the learning rate

```python
# Create a Learning Rate Scheduler, which will test a different LR at each epoch
lr_schedule = tf.keras.callbacks.LearningRateScheduler(
    lambda epoch: 1e-8 * 10**(epoch / 20))
```

```python
# Initialize the optimizer
optimizer = tf.keras.optimizers.Adam()

# Set the training parameters
model = lstm_model(X_train)
model.compile(loss='mean_squared_error', optimizer=optimizer)
```

```python
# Train the model
history = model.fit(X_train, y_train, batch_size=32, epochs=100, validation_split=0.2, callbacks=[lr_schedule])
```

```python
# Define the learning rate array
lrs = 1e-8 * (10 ** (np.arange(100) / 20))

# Set the figure size
plt.figure(figsize=(10, 6))

# Set the grid
plt.grid(True)

# Plot the loss in log scale
plt.semilogx(lrs, history.history["loss"])

# Increase the tickmarks size
plt.tick_params('both', length=10, width=1, which='both')

# Set the plot boundaries
plt.axis([1e-8, 1e-3, 0, .3])
```

Here it shows the learning rate continues to decline until 10^3, so 10^-3 was used as the optimal learning rate.
