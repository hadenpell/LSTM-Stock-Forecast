```python
import pandas as pd
import numpy as np
from pandas_datareader.data import DataReader
import yfinance as yf
from pandas_datareader import data as pdr
import seaborn as sns
from prophet import Prophet
import matplotlib.pyplot as plt
from datetime import datetime
import plotly.express as px
from sklearn.metrics import mean_squared_error

yf.pdr_override()
```

```python
stock_list = ['MAT','HEAR','OTLY']
```

```python
dfs={}
```

```python
# Get 2 years of data
end = datetime.now()
start = datetime(end.year - 2, end.month, end.day)

for stock in stock_list:
    dfs[stock] = yf.download(stock, start, end)
```

```python
# Decide on sizes for training/test sets
# Predict 60 days
test_size=60 
```

```python
# Split dfs into train and test
train_dfs,test_dfs={},{}

for stock,df in dfs.items():
    train_size=df.shape[0]-test_size
    print(train_size,test_size)
    train_dfs[stock],test_dfs[stock]=df.iloc[0:train_size,:],df.iloc[train_size - test_size: , :]
```

### Baseline Moving Average Model

```python
# Run a MA model for each stock and report RMSE
for stock,df in train_dfs.items():
    train_data=df.filter(['Close'])
    test_data = test_dfs[stock].filter(['Close'])
    train_ma = train_data.rolling(window=60).mean().dropna()
    test_predictions = np.repeat(train_ma.iloc[-1], len(test_data)) # Repeat the last moving average value
    rmse = np.sqrt(mean_squared_error(test_data, test_predictions))
    print(f'{stock} RMSE: {rmse:.2f}')
```

### Prophet Time Series Model

```python
proph_models,proph_predictions={},{}
```

```python
# Train model, make predictions and report RMSE
for stock,df in train_dfs.items():
    # Format df for Prophet model
    train_df=df.filter(['Close']).reset_index().rename(columns={'Date':'ds','Close':'y'})
    test_df=test_dfs[stock].filter(['Close']).reset_index().rename(columns={'Date':'ds','Close':'y'})
    # Fit model
    m=Prophet()
    m.fit(train_df)
    proph_models[stock]=m
    predicted_test=m.predict(test_df)[['yhat']]
    actual_test=test_dfs[stock].filter(['Close'])
    proph_predictions[stock]=[actual_test,predicted_test]
    rmse = np.sqrt(mean_squared_error(actual_test, predicted_test))
    print(f'{stock} RMSE: {rmse:.2f}')
```

```python
# Plot Prophet test set predictions for each stock
for stock,preds in proph_predictions.items():
    actual,predicted=preds[0],preds[1]
    plt.figure(figsize=(16,6))
    plt.title(f'{stock} Prophet Model Test Set Predictions')
    plt.xlabel('Date', fontsize=18)
    plt.ylabel('Close Price USD ($)', fontsize=18)
    plt.plot(actual.reset_index()['Close'])
    plt.plot(predicted.reset_index()['yhat'])
```

### LSTM model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
from sklearn.preprocessing import MinMaxScaler
```

```python
# Set global variables for LSTM model

# Size of test set
PRED_DAYS=60
# Size of sequence for forecasting
SEQ_DAYS=60
# Days to forecast ahead
FCST_DAYS=30
```

```python
def lstm_model(x_train):
    print(x_train.shape)
    # Build the LSTM model
    model = Sequential()
    model.add(LSTM(128, return_sequences=True, input_shape= (x_train.shape[1], 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(64, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(25))
    model.add(Dense(1))
    return model
```

```python
preds={}
models={}
```

```python
for stock, df in dfs.items():
    data = dfs[stock].filter(['Close'])
    dataset = data.values

    # Normalize data
    scaler = MinMaxScaler(feature_range=(0,1))
    scaled_data = scaler.fit_transform(data['Close'].values.reshape(-1,1))

    # Set the number of days used for prediction
    #prediction_days = 60

    # Initialize empty lists for training data input and output
    x_train = []
    y_train = []

    # Iterate through the scaled data, starting from the PRED_DAYS index
    for x in range(PRED_DAYS, len(scaled_data)):
        # Append the previous 'PRED_DAYS' values to x_train
        x_train.append(scaled_data[x - PRED_DAYS:x, 0])
        # Append the current value to y_train
        y_train.append(scaled_data[x, 0])

    # Convert the x_train and y_train lists to numpy arrays
    x_train, y_train = np.array(x_train), np.array(y_train)

    # Reshape x_train to a 3D array with the appropriate dimensions for the LSTM model
    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
    
    model = lstm_model(x_train)

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    model.fit(x_train, y_train, batch_size=1, epochs=10)
    models[stock]=model
    
    # Get the number of rows to train the model on
    training_data_len = data.shape[0] -  PRED_DAYS
    
    test_data = scaled_data[training_data_len - PRED_DAYS: , :]
    # Create the data sets x_test and y_test
    x_test = []
    y_test = dataset[training_data_len:, :]
    
    for i in range(PRED_DAYS, len(test_data)):
        x_test.append(test_data[i-PRED_DAYS:i, 0])

    # Convert the data to a numpy array
    x_test = np.array(x_test)

    # Reshape the data
    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1 ))

    # Get the models predicted price values 
    predictions = model.predict(x_test)
    predictions = scaler.inverse_transform(predictions)
    
    # Store the actuals for train set, predictions for test set, and actuals for test set
    preds[stock]=[data[:training_data_len],predictions,data[training_data_len:]]
    
    # Get the root mean squared error (RMSE)
    rmse = np.sqrt(np.mean(((predictions - y_test) ** 2)))
    print(f'{stock} RMSE: {rmse:.2f}')
```

```python
# Visualize LSTM predictions
for stock,p in preds.items():
    train,predicted,test=p[0],p[1],p[2]
    test['Predicted']=predicted
    plt.figure(figsize=(16,6))
    plt.title(f'{stock} LSTM Model Test Set Predictions')
    plt.xlabel('Date', fontsize=18)
    plt.ylabel('Close Price USD ($)', fontsize=18)
    plt.plot(test[['Close','Predicted']])
    plt.plot(train['Close'])
```

### Forecasting 30 days
https://stackoverflow.com/questions/69906416/forecast-future-values-with-lstm-in-python

```python
SEQ_AMT=60
FCST_AMT=30
```

```python
models
```

```python
forecasts={}
```

```python
# Loop
```

```python
for stock,df in dfs.items():
    data = dfs[stock].filter(['Close'])
    model = models[stock]
    
    # Input sequence should contain the most recent 60 data points
    input_sequence = data[-SEQ_AMT:].Close.values

    # Initialize the scaler
    scaler = MinMaxScaler(feature_range=(0, 1))

    # Scale the input sequence
    input_sequence = scaler.fit_transform(input_sequence.reshape(-1, 1))

    # Reshape the input sequence to match the LSTM input shape
    input_sequence = np.reshape(input_sequence, (input_sequence.shape[0], input_sequence.shape[1], 1))

    forecasted_values = []  # To store the forecasted values

    for _ in range(FCST_AMT):  # Forecast next 30 days
        # Predict the next data point (one step ahead)
        next_value = model.predict(input_sequence)

        # Append the predicted value to the forecasted values
        forecasted_values.append(next_value[0, 0])

        # Update the input sequence for the next iteration
        input_sequence = np.append(input_sequence, next_value.reshape(input_sequence.shape[0], 1, 1), axis=1)
        
    forecasted_values = scaler.inverse_transform(np.array(forecasted_values).reshape(-1,1))
    df_future = pd.DataFrame(columns=['Date', 'Close', 'Forecast'])
    df_future['Date'] = pd.date_range(start=data.reset_index().Date.iloc[-1] + pd.Timedelta(days=1), periods=FCST_AMT)
    df_future['Forecast'] = forecasted_values
    results = data.append(df_future.set_index('Date'))
    forecasts[stock]=results
```

```python
import plotly.graph_objects as go
```

```python
for stock,p in preds.items():
    train,predicted,test=p[0],p[1],p[2]
    results=forecasts[stock]
    test['Predicted']=predicted
    
    # Create a figure
    fig = go.Figure()

    # Plot the test predictions
    fig.add_trace(go.Scatter(x=test.index, y=test['Close'], mode='lines', name='Test Predictions'))
    fig.add_trace(go.Scatter(x=test.index, y=test['Predicted'], mode='lines', name='Test Data'))

    # Plot the train data
    fig.add_trace(go.Scatter(x=train.index, y=train['Close'], mode='lines', name='Train Data'))

    # Plot the initial forecast
    fig.add_trace(go.Scatter(x=results.index, y=results['Forecast'], mode='lines', name='30 Day Forecast'))
    
    fig.update_layout(title="Connected Line Chart", xaxis_title="Date", yaxis_title="Value")

    fig.show()
```
